```python
data = pd.DataFrame({
    'x0': [1, 2, 3, 4, 5],
    'x1': [0.01, -0.01, 0.25, -4.1, 0.],
    'y': [-1.5, 0., 3.6, 1.3, -2.]})
data
```

- Importowanie bibliotek: Importujesz Pandas oraz `patsy`.
- Tworzenie przykładowych danych: Tworzysz DataFrame `data`, który zawiera kolumny `y`, `x0` i `x1`.
- Tworzenie macierzy projektowych: Używasz funkcji `patsy.dmatrices()`, aby utworzyć macierz dla zmiennej zależnej `y` oraz macierz projektową dla zmiennych niezależnych (`x0` i `x1`). Wzór `'y ~ x0 + x1'` określa, że `y` jest zmienną zależną, a `x0` i `x1` są zmiennymi niezależnymi.

```python
import patsy
y, X = patsy.dmatrices('y ~ x0 + x1', data)
y
```

```python
X
```

```python
np.asarray(y)
```

```python
np.asarray(X)
```

```python
patsy.dmatrices('y ~ x0 + x1 + 0', data)[1]
```

```python
coef, resid, _, _ = np.linalg.lstsq(X, y)
```

```python
coef
coef = pd.Series(coef.squeeze(), index=X.design_info.column_names)
coef
```

```python
y, X = patsy.dmatrices('y ~ x0 + np.log(np.abs(x1) + 1)', data)
X
```

- Wzór `'v2 ~ key1'` oznacza, że zmienna zależna to `v2`, a zmienną niezależną jest `key1`, która jest zmienną kategoryczną. `patsy` automatycznie przekształca zmienną kategoryczną na zmienne dummy.

```python
data = pd.DataFrame({
    'key1': ['a', 'a', 'b', 'b', 'a', 'b', 'a', 'b'],
    'key2': [0, 1, 0, 1, 0, 1, 0, 0],
    'v1': [1, 2, 3, 4, 5, 6, 7, 8],
    'v2': [-1, 0, 2.5, -0.5, 4.0, -1.2, 0.2, -1.7]
})
y, X = patsy.dmatrices('v2 ~ key1', data)
X
```

```python
y, X = patsy.dmatrices('v2 ~ key1 + 0', data)
X
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
rng = np.random.default_rng(seed=12345)

def dnorm(mean, variance, size=1):
    if isinstance(size, int):
        size = size,
    return mean + np.sqrt(variance) * rng.standard_normal(*size)

N = 100
X = np.c_[dnorm(0, 0.4, size=N),
          dnorm(0, 0.6, size=N),
          dnorm(0, 0.2, size=N)]
eps = dnorm(0, 0.1, size=N)
beta = [0.1, 0.3, 0.5]

y = np.dot(X, beta) + eps
```

```python
X[:5]
y[:5]
```

```python
X_model = sm.add_constant(X)
X_model[:5]
```

```python
model = sm.OLS(y, X)
```

```python
results = model.fit()
results.params
```

```python
print(results.summary())
```

```python
data = pd.DataFrame(X, columns=['col0', 'col1', 'col2'])
data['y'] = y
data[:5]
```

```python
results = smf.ols('y ~ col0 + col1 + col2', data=data).fit()
results.params
results.tvalues
```

```python
results.predict(data[:5])
```

```python
init_x = 4

values = [init_x, init_x]
N = 1000

b0 = 0.8
b1 = -0.4
noise = dnorm(0, 0.1, N)
for i in range(N):
    new_x = values[-1] * b0 + values[-2] * b1 + noise[i]
    values.append(new_x)
```

```python
from statsmodels.tsa.ar_model import AutoReg
MAXLAGS = 5
model = AutoReg(values, MAXLAGS)
results = model.fit()
```

```python
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
train.head(4)
```

```python
train.isna().sum()
test.isna().sum()
```

```python
impute_value = train['Age'].median()
train['Age'] = train['Age'].fillna(impute_value)
test['Age'] = test['Age'].fillna(impute_value)
```

```python
train['IsFemale'] = (train['Sex'] == 'female').astype(int)
test['IsFemale'] = (test['Sex'] == 'female').astype(int)
```

```python
predictors = ['Pclass', 'IsFemale', 'Age']

X_train = train[predictors].to_numpy()
X_test = test[predictors].to_numpy()
y_train = train['Survived'].to_numpy()
X_train[:5]
y_train[:5]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
y_predict[:10]
```

```python
from sklearn.linear_model import LogisticRegressionCV
model_cv = LogisticRegressionCV(Cs=10)
model_cv.fit(X_train, y_train)
```
