```python
from datetime import datetime

fig, ax = plt.subplots()

data = pd.read_csv("/content/spx.csv", index_col=0, parse_dates=True)
spx = data["SPX"]

spx.plot(ax=ax, color="black")

crisis_data = [
    (datetime(2007, 10, 11), "Szczyt hossy"),
    (datetime(2008, 3, 12), "Upadek Bear Stearns"),
    (datetime(2008, 9, 15), "Bankructwo Lehman")
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor="black", headwidth=4, width=2,
                                headlength=4),
                horizontalalignment="left", verticalalignment="top")

# Przedział lat 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])
ax.set_xlabel("Data")

ax.set_title("Ważne wydarzenia kryzysu finansowego 2008-2009")
```

- `np.log(data)`: Zastosowanie logarytmu naturalnego do każdej wartości w DataFrame. Jest to powszechna technika stosowana w analizie szeregów czasowych, która może pomóc w stabilizacji wariancji.
- `.diff()`: Oblicza różnice między kolejnymi wartościami w każdej kolumnie. Ta operacja pozwala na uzyskanie zmian (przyrostów) w logarytmicznych wartościach.
`.dropna()`: Usuwa wszelkie wiersze, które zawierają wartości `NaN`, które mogą powstać na początku DataFrame po zastosowaniu `.diff()`.

```python
macro = pd.read_csv("/content/macrodata.csv")
data = macro[["cpi", "m1", "tbilrate", "unemp"]]
trans_data = np.log(data).diff().dropna()
trans_data.tail()
```

```python
plt.figure()
```

```python
ax = sns.regplot(x="m1", y="unemp", data=trans_data)
plt.title("Zmiany wartości log(m1) w porównaniu z log(unemp)")
```

```python
sns.pairplot(trans_data, diag_kind="kde", plot_kws={"alpha": 0.2})
```
