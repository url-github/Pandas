```python
frame = pd.read_csv('/content/Mcdonalds.csv')
frame.head()
```

Do celów ćwiczeń najpierw brudzę w danych w celu późniejszego oczyszczenia

```python
frame.loc[1,'Calories']=np.NaN
frame.loc[2,'TotalFat']=np.NaN
frame.head()
```

Wiersze z NaN poprostu zniknęły

```python
frame.dropna().head()
```

Ta instrukcja usunie tylko te wiersze w których wszędzie jest NaN

```python
frame.dropna(how='all').head()
```

Parametrem subset określam, które kolumny mają być zweryfikowane przez dropna

```python
frame.dropna(subset=['TotalFat'], how='any').head()
```

Tym poleceniem usunę wszystkie kolumny w których wystąpił NaN

```python
frame.dropna(axis='columns').head()
```

Do tej pory pracowałem na kopii danych. Jeśli chcę utworzyć zmianę na stałę do dodaje parametr inplace=True (domyślnie to False)

```python
frame.head()
```

```python
frame.dropna(inplace=True)
frame.head()
```
