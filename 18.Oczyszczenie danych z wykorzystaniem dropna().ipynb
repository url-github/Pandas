Do celw wicze najpierw brudz w danych w celu pniejszego oczyszczenia

```python
frame.loc[1,'Calories']=np.NaN
frame.loc[2,'TotalFat']=np.NaN
frame.head()
```

Wiersze z NaN poprostu znikny

```python
frame.dropna().head()
```

Ta instrukcja usunie tylko te wiersze w ktrych wszdzie jest NaN

```python
frame.dropna(how='all').head()
```

Parametrem subset okrelam, ktre kolumny maj by zweryfikowane przez dropna

```python
frame.dropna(subset=['TotalFat'], how='any').head()
```

Tym poleceniem usun wszystkie kolumny w ktrych wystpi NaN

```python
frame.dropna(axis='columns').head()
```
