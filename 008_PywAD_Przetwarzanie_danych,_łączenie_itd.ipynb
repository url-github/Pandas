```python
data = pd.Series(
    np.random.rand(6),
    index=[['A', 'A', 'B', 'B', 'C', 'C'], ['x', 'y', 'x', 'y', 'x', 'y'], [1, 2, 1, 2, 1, 2]]
)

print("Oryginalny obiekt:")
print(data)

# Rozpakowanie drugiego poziomu (x, y)
unstacked = data.unstack(level=1)
print("\nPo zastosowaniu unstack(level=1):")
print(unstacked)
```

- Funkcja `pd.merge()` łączy oba DataFrame na podstawie kolumn `key1` i `key2`. Używając parametru `how="outer"`, wykonujesz pełne złączenie zewnętrzne (full outer join), co oznacza, że wszystkie wiersze z obu DataFrame będą zachowane.

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})

right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})

pd.merge(left, right, on=["key1", "key2"], how="outer")
```

```python
pd.merge(left, right, on="key1")
```

```python
pd.merge(left, right, on="key1", suffixes=("_left", "_right"))
```

```python
lefth = pd.DataFrame({"key1": ["Ohio", "Ohio", "Ohio",
                               "Nevada", "Nevada"],
                      "key2": [2000, 2001, 2002, 2001, 2002],
                      "data": pd.Series(range(5), dtype="Int64")})

righth_index = pd.MultiIndex.from_arrays(
    [
        ["Nevada", "Nevada", "Ohio", "Ohio", "Ohio", "Ohio"],
        [2001, 2000, 2000, 2000, 2001, 2002]
    ]
)
righth = pd.DataFrame({"event1": pd.Series([0, 2, 4, 6, 8, 10], dtype="Int64",
                                           index=righth_index),
                       "event2": pd.Series([1, 3, 5, 7, 9, 11], dtype="Int64",
                                           index=righth_index)})
print(lefth)
print(righth)
```

- left_on=["key1", "key2"]: Określa, że kolumny key1 i key2 w lefth będą używane jako klucze do łączenia.
- right_index=True: Określa, że indeksy w righth będą używane jako klucze do łączenia.

```python
pd.merge(lefth, righth, left_on=["key1", "key2"], right_index=True)
```

- left_on=["key1", "key2"]: Klucze w DataFrame lefth to kombinacja kolumn key1 i key2.
- right_index=True: Indeksy w DataFrame righth będą używane jako klucze do łączenia.
- how="outer": Włącza w wynik wszystkie wiersze zarówno z lefth, jak i righth, wypełniając wartości brakujące (NaN) dla tych, które nie mają dopasowania.

```python
pd.merge(lefth, righth, left_on=["key1", "key2"],
         right_index=True, how="outer")
```

- Indeks jest używany do identyfikacji i dostępu do danych w obrębie jednego obiektu Pandas (Series lub DataFrame), podczas gdy klucze są stosowane podczas operacji łączenia lub konkatenacji, aby oznaczyć źródła danych w wyniku.
- Indeks może być zmieniany i ustawiany na różne kolumny, a klucze są definiowane podczas operacji łączenia, co pozwala na lepszą organizację wynikowych danych.

- `names="upper", "lower"`: Umożliwia nadanie nazw poziomom MultiIndex. W tym przypadku `upper` będzie nazwą górnego poziomu (kluczy), a `lower` będzie nazwą dolnego poziomu (indeksów oryginalnych DataFrame).

```python
pd.concat([df1, df2], axis="columns", keys=["level1", "level2"],
          names=["upper", "lower"])
```

- Funkcja `pd.isna()` zwraca maskę booleanową, która wskazuje, które elementy w tablicy `a` są `NaN`. Zwraca ona tablicę o tym samym kształcie co `a`, gdzie dla każdego elementu, który jest `NaN`, wartość to `True`, a dla pozostałych elementów to `False`.

- Cała konstrukcja `np.where(pd.isna(a), b, a)` oznacza: “Zastąp wartości `NaN` w tablicy `a` wartościami z tablicy `b`, a dla pozostałych wartości pozostaw oryginalne wartości z tablicy `a`”.

```python
np.where(pd.isna(a), b, a)
```

- Funkcja `pd.melt()` jest przydatnym narzędziem do przekształcania DataFrame z formatu szerokiego na format długi. Umożliwia to lepszą organizację danych do analizy i wizualizacji. W podanym przykładzie kolumna `"key"` została zachowana jako identyfikator, a pozostałe kolumny zostały roztopione w nowym formacie.

```python
melted = pd.melt(df, id_vars="key")
melted
```

```python
reshaped = melted.pivot(index="key", columns="variable",
                        values="value")reshaped
```

```python
reshaped.reset_index()
```

- `id_vars`: Kolumna (lub kolumny), które mają być używane jako identyfikatory. W tym przypadku jest to kolumna `"key"`, która zostanie zachowana w wyniku.
- `value_vars`: Kolumny, które mają być “roztopione”. W tym przypadku są to kolumny `"A"` i `"B"`. Jeśli ten parametr nie zostanie określony, Pandas użyje wszystkich pozostałych kolumn poza tymi wymienionymi w `id_vars`.

```python
pd.melt(df, id_vars="key", value_vars=["A", "B"])
```

```python
pd.melt(df, value_vars=["A", "B", "C"])
```

```python
pd.melt(df, value_vars=["key", "A", "B"])
```
