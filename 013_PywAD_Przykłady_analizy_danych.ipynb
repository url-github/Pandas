```python
def get_counts(sequence):
    counts = {}
    for x in sequence:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1
    return counts
```

```python
from collections import Counter

def get_counts2(sequence):
    return Counter(sequence)
```

```python
def top_counts(count_dict, n=10):
    # Tworzymy listę par (licznik, klucz)
    value_key_pairs = [(count, tz) for tz, count in count_dict.items()]
    # value_key_pairs = [[count, tz] for tz, count in count_dict.items()]

    # Sortujemy w odwrotnej kolejności
    value_key_pairs.sort(reverse=True)

    # Zwracamy n największych elementów
    return value_key_pairs[:n]
```

size() - to metoda, która zwraca liczbę elementów w każdej grupie.

unstack() - to metoda, która przekształca serię w DataFrame, przesuwając jeden z poziomów indeksu (najczęściej drugi poziom) do kolumn.

```python
agg_counts = by_tz_os.size().unstack().fillna(0) # fillna zastępuje N/A na 0
agg_counts.head()
```

- Metoda `argsort()` zwraca indeksy posortowanej wersji serii. Oznacza to, że zwróci tablicę indeksów, które wskazują, w jakiej kolejności elementy byłyby posortowane.

- Metoda `take()` służy do pobierania wierszy (lub kolumn) z DataFrame na podstawie podanych indeksów. W tym przypadku, `take(indexer-10:)` zwraca ostatnie 10 kolumn z DataFrame `agg_counts`, które mają największe sumy.

```python
count_subset = agg_counts.take(indexer[-10:])
count_subset
```

- Metoda `reset_index()` w bibliotece Pandas jest używana do resetowania indeksu DataFrame do domyślnego indeksu numerycznego (0, 1, 2, …).


```python
count_subset = count_subset.stack()
count_subset.name = "total"
count_subset = count_subset.reset_index()
count_subset.head(10)
gfg = sns.barplot(x="total", y="tz", hue="os",  data=count_subset)
gfg.set(xlabel ="Liczba", ylabel = "Strefa czasowa")
plt.legend(title='System operacyjny')
```

```python
pieces = []
for year in range(2008, 2011):
    path = f"/content/yob{year}.txt"
    frame = pd.read_csv(path, names=["name", "sex", "births"])

    # Dodaj kolumnę dla danego roku
    frame["year"] = year
    pieces.append(frame)
```

```python
pieces
```

- `pieces`: To lista lub inny iterowalny obiekt, który zawiera DataFrame’y (lub serie), które chcesz połączyć w jeden. Każdy element w `pieces` powinien mieć tę samą strukturę kolumn (tj. te same nazwy kolumn).
- `pd.concat()`: Funkcja ta łączy wszystkie DataFrame’y w `pieces` w jeden DataFrame.
- `ignore_index=True`: Ustawienie tego argumentu na `True` sprawia, że nowy DataFrame otrzyma nowy, ciągły indeks (0, 1, 2, …)

```python
names = pd.concat(pieces, ignore_index=True)
names
```

- `pivot_table()`: Metoda ta tworzy tabelę przestawną, która agreguje dane w formie tabeli.
- `"births"`: To kolumna, której wartości będą agregowane (liczba urodzeń).
- `index="year"`: Ustala, że wiersze tabeli przestawnej będą reprezentować lata.
- `columns="sex"`: Ustala, że kolumny tabeli przestawnej będą reprezentować płeć (np. “Male” i “Female”).
- `aggfunc=sum`: Określa funkcję agregującą, która w tym przypadku oblicza sumę urodzeń dla każdego roku i płci.

```python
total_births = names.pivot_table("births", index="year",
                                 columns="sex", aggfunc=sum)
total_births.tail()
```

```python
gfg = total_births.plot(title="Całkowita liczba urodzeń z podziałem na płeć i lata")
gfg.set(xlabel ="Rok", ylabel = "Liczba")
plt.legend(title="Płeć", labels=["dziewczynki", "chłopcy"])
```

- Argument `group`: To podgrupa DataFrame, która jest przekazywana do funkcji przez metodę `apply()`. Zawiera wszystkie wiersze dla danej kombinacji roku i płci.
- Obliczanie proporcji: Wewnątrz funkcji dodajesz nową kolumnę `"prop"` do grupy, która jest obliczana jako stosunek liczby urodzeń w danej grupie (kolumna `"births"`) do sumy urodzeń w tej samej grupie.
- Zwracanie grupy: Funkcja zwraca zmodyfikowaną grupę, która teraz zawiera dodatkową kolumnę z proporcjami.

```python
def add_prop(group):
    group["prop"] = group["births"] / group["births"].sum()
    return group
names = names.groupby(["year", "sex"]).apply(add_prop)
```

- `cumsum()`: Metoda ta oblicza skumulowaną sumę wartości w posortowanej kolumnie, co oznacza, że dla każdego elementu zwraca sumę wszystkich poprzednich elementów oraz samego siebie.

- `searchsorted(0.5)`: Metoda ta znajduje indeks, w którym wartość 0.5 mogłaby być wstawiona w posortowanej tablicy `prop_cumsum`, aby zachować porządek. Innymi słowy, zwraca pierwszy indeks, dla którego skumulowana suma jest większa lub równa 0.5.

```python
prop_cumsum = df["prop"].sort_values(ascending=False).cumsum()
prop_cumsum[:10]
```

```python
prop_cumsum.searchsorted(0.5)
```

- `sort_values("prop", ascending=False)`: Sortuje wartości w kolumnie `prop` (proporcje) w kolejności malejącej.

- `.prop.cumsum()`: Oblicza skumulowaną sumę wartości w posortowanej kolumnie `prop`.

- `searchsorted(0.5)`: Metoda ta znajduje indeks, w którym wartość 0.5 mogłaby być wstawiona w posortowanej tablicy `in1900`, aby zachować porządek.

```python
df = boys[boys.year == 2009]
in1900 = df.sort_values("prop", ascending=False).prop.cumsum()
in1900.searchsorted(0.5) + 1
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values("prop", ascending=False)
    return group.prop.cumsum().searchsorted(q) + 1

diversity = top1000.groupby(["year", "sex"]).apply(get_quantile_count)
diversity = diversity.unstack()
```

- `table.sum()`: Ta część oblicza sumę wszystkich wartości w DataFrame `table`.
- `table / table.sum()`: Dzieląc każdy element w `table` przez sumę, uzyskujemy proporcje dla każdej litery. Wynikiem będzie nowy DataFrame - `letter_prop`, gdzie każda wartość reprezentuje udział danej litery w całości.
- `letter_prop.loc["d", "n", "y", "M"]`: Używając metody `loc`, wyciągamy dane dla liter `"d"`, `"n"` i `"y"` w kolumnie `"M"`. Zakładamy, że kolumna `"M"` istnieje w DataFrame `letter_prop`.
- `.T`: Transponuje wynik, co oznacza, że wiersze stają się kolumnami i odwrotnie.

```python
letter_prop = table / table.sum()

dny_ts = letter_prop.loc[["d", "n", "y"], "M"].T
dny_ts.head()
```
