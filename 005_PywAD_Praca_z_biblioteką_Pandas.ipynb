- 	`fill_value`: Wartość używana do wypełnienia istniejących brakujących (NaN) wartości oraz wszelkich nowych elementów potrzebnych do udanej alokacji DataFrame. Jeśli w obu odpowiadających lokalizacjach DataFrame’ów dane są brakujące, wynik również będzie brakujący.

```python
df1.add(df2, fill_value=0)
```

```python
df1
```

```python
1 / df1
```

- W bibliotece pandas metoda `rdiv()` służy do wykonywania operacji dzielenia, gdzie obiekt (np. DataFrame lub Series) jest dzielony przez wartość z prawej strony.

```python
df1.rdiv(1)
```

- Wykonanie polecenia `df1.reindex(columns=df2.columns, fill_value=0)` w bibliotece pandas służy do zmiany etykiet kolumn w DataFrame `df1` na takie, które odpowiadają etykietom kolumn w DataFrame `df2`.
- W przypadku, gdy niektóre kolumny z `df2` nie istnieją w `df1`, zostaną one dodane z wartością `0` wypełniającą brakujące dane.

```python
df1.reindex(columns=df2.columns, fill_value=0)
```
