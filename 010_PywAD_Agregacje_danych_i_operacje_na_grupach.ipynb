```python
frame = pd.DataFrame({"dane1": np.random.standard_normal(1000),
                      "dane2": np.random.standard_normal(1000)})
frame.head()
```

```python
quartiles = pd.cut(frame["dane1"], 4)
quartiles.head(10)
```

```python
def get_stats(group):
    return pd.DataFrame(
        {"min": group.min(), "max": group.max(),
        "count": group.count(), "mean": group.mean()}
    )

grouped = frame.groupby(quartiles)
grouped.apply(get_stats)
```

```python
print(grouped.agg(["min", "max", "count", "mean"]))
```

- `pd.qcut()`: Funkcja ta dzieli dane na rwne czci (kwantyle). W tym przypadku, argument `4` oznacza, e dane bd podzielone na cztery kwantyle (kwarty).
- `labels=False`: Ustawienie tego parametru na `False` powoduje, e funkcja zwraca numery kwantyli zamiast etykiet. Oznacza to, e wartoci bd reprezentowane jako liczby cakowite od 0 do 3

```python
quartiles_samp = pd.qcut(frame["dane1"], 4, labels=False)
quartiles_samp.head()
```

```python
grouped = frame.groupby(quartiles_samp)
grouped.apply(get_stats)
```

- Metoda `corrwith()` oblicza wspczynnik korelacji pomidzy kolumn `"SPX"` a wszystkimi innymi kolumnami w tej grupie. Wynikiem jest Series, w ktrym indeksy to nazwy kolumn, a wartoci to wspczynniki korelacji.

```python
def spx_corr(group):
    return group.corrwith(group["SPX"])
```

```python
rets = close_px.pct_change().dropna()
```

```python
def get_year(x):
    return x.year

by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
def corr_aapl_msft(group):
    return group["AAPL"].corr(group["MSFT"])
by_year.apply(corr_aapl_msft)
```

```python
import statsmodels.api as sm
def regress(data, yvar=None, xvars=None):
    Y = data[yvar]
    X = data[xvars]
    X["intercept"] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
#! ipython id=78fdb9df1f984f3cb2ae8ae664af69f9
by_year.apply(regress, yvar="AAPL", xvars=["SPX"])
```
