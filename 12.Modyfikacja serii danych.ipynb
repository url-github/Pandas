```python
# Definiuje funkcje aby podmieni wybrane wartoci w serii

def ReplaceType(oldType):
  if oldType == "Grass" or oldType == "Ground":
    return "Nature"
  else:
    return oldType
```

```python
ReplaceType('Fire')
```

```python
ReplaceType('Grass')
```

```python
ReplaceType('Ground')
```

```python
ReplaceType('Nature')
```

```python
ReplaceType('oldType')
```

```python
ReplaceType('example')
```

```python
# Kiedy mam do zaaplikowania funkcj do serii do uywam metody apply

pokNature = pokString.apply(ReplaceType)
pokNature.head()
```
