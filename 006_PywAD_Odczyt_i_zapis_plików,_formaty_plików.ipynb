- `chunksize=1000`: Ten parametr okrela, e dane maj by wczytywane w kawakach po 1000 wierszy. Zamiast adowa cay plik do pamici, pandas zwraca obiekt `TextFileReader`, ktry jest iteratorem. Mona go uywa do przetwarzania danych w mniejszych partiach.

```python
chunker = pd.read_csv("/content/ex6.csv", chunksize=1000)
type(chunker)
```

```python
chunker
```

```python
tot = pd.Series([], dtype='int64')
for piece in chunker:
    tot = tot.add(piece["key"].value_counts(), fill_value=0)

tot = tot.sort_values(ascending=False)
```

```python
# Inicjalizacja pustej serii do przechowywania sum
tot = pd.Series([], dtype='int64')

# Przetwarzanie danych kawaek po kawaku
for piece in chunker:
    # 1. piece["key"].value_counts() - zlicza wystpienia wartoci w kolumnie "key"
    # 2. add(..., fill_value=0) - dodaje zliczenia do tot, uywajc 0 dla brakujcych wartoci
    tot = tot.add(piece["key"].value_counts(), fill_value=0)

# Sortowanie kocowych wynikw malejco
tot = tot.sort_values(ascending=False)
```

```python
from lxml import objectify

# cieka do pliku XML
path = "/content/Performance_MNR.xml"

# Otwieramy i parsujemy plik XML
with open(path) as f:
    parsed = objectify.parse(f)

# Pobieramy gwny element (root) dokumentu XML
root = parsed.getroot()
```

```python
root.INDICATOR
```

```python
data = []

skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
               "DESIRED_CHANGE", "DECIMAL_PLACES"]

for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

```python
# Lista na przetworzone dane
data = []

# Pola, ktre chcemy pomin podczas przetwarzania
skip_fields = ["PARENT_SEQ", "INDICATOR_SEQ",
               "DESIRED_CHANGE", "DECIMAL_PLACES"]

# Iteracja po wszystkich elementach INDICATOR w root
for elt in root.INDICATOR:
    # Sownik na dane z jednego elementu
    el_data = {}

    # Iteracja po wszystkich dzieciach elementu INDICATOR
    for child in elt.getchildren():
        # Pomijamy pola zdefiniowane w skip_fields
        if child.tag in skip_fields:
            continue
        # Zapisujemy warto elementu (pyval konwertuje do odpowiedniego typu Pythona)
        el_data[child.tag] = child.pyval

    # Dodajemy przetworzony element do listy
    data.append(el_data)
```

```python
# Tworzenie bazy SQLite i tabeli
import sqlite3

# Definicja struktury tabeli
query = """
CREATE TABLE test
(a VARCHAR(20), b VARCHAR(20),
c REAL,        d INTEGER
);"""

# Poczenie z baz i utworzenie tabeli
con = sqlite3.connect("mydata.sqlite")
con.execute(query)
con.commit() # Zatwierdzenie zmian

# Struktura tabeli:
# - a: tekst do 20 znakw
# - b: tekst do 20 znakw
# - c: liczba zmiennoprzecinkowa
# - d: liczba cakowita
```

```python
data = [("Atlanta", "Georgia", 1.25, 6),
        ("Tallahassee", "Florida", 2.6, 3),
        ("Sacramento", "California", 1.7, 5)]
stmt = "INSERT INTO test VALUES(?, ?, ?, ?)"

con.executemany(stmt, data)
con.commit()
```
